```python
from sklearn.preprocessing import StandardScaler
```

```python
def saran (data):
    stand = StandardScaler()
    final = stand.fit_transform(data)
    
    return final
```

```python
X_train = pd.DataFrame(saran(x_train),  columns=['Gender', 'Age', 'Occupation',
       'Stay_In_Current_City_Years', 'Marital_Status', 'cat_1', 'cat_2',
       'cat_3', 'B', 'C'])
```
